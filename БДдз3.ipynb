```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import seaborn as sns
```

```python
data = pd.read_csv('/content/insurance.csv')
```

```python
data.describe()
```

```python
data.hist(color = 'g', edgecolor = 'black')
plt.show()
```

```python
mean_charges = np.mean(data['charges'])
moda_charges = sts.mode(data['charges'])
med_charges = np.median(data['charges'])

mean_bmi = np.mean(data['bmi'])
moda_bmi = sts.mode(data['bmi'])
med_bmi = np.median(data['bmi'])

std_charges = data['charges'].std()
std_bmi = data['bmi'].std()

print('Среднее charges =', mean_charges)
print('Мода charges:', moda_charges)
print('Медиана charges =', med_charges)
print('Стандартное отклонение charges:', std_charges)
print()
print('Среднее bmi =', mean_bmi)
print('Мода bmi:', moda_bmi)
print('Медиана bmi =', med_bmi)
print('Стандартное отклонение bmi:', std_bmi)
```

```python
val = [mean_bmi,moda_bmi[0],med_bmi,std_bmi]
name = ['Среднее', 'Мода', 'Медиана', 'Стандартное отклонение']
data['bmi'].hist(color = 'g', edgecolor = 'black')

plt.axvline(x = val[0], color = 'r', label = name[0])
plt.axvline(x = val[1], color = 'b', label = name[1])
plt.axvline(x = val[2], color = 'c', label = name[2])
plt.axvline(x = val[3], color = 'y', label = name[3])

# place legend outside
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
plt.title('bmi')
plt.show()
```

```python
val = [mean_charges,moda_charges[0],med_charges,std_charges]
data['charges'].hist(color = 'g', edgecolor = 'black')

plt.axvline(x = val[0], color = 'r', label = name[0])
plt.axvline(x = val[1], color = 'b', label = name[1])
plt.axvline(x = val[2], color = 'c', label = name[2])
plt.axvline(x = val[3], color = 'y', label = name[3])

# place legend outside
plt.legend(bbox_to_anchor = (1.0, 1), loc = 'upper left')
plt.title('charges')
plt.show()
```

```python
plt.boxplot([data['age'],data['bmi']],labels=['age','bmi'],vert = False)
plt.xticks(np.arange(10,70,5))
plt.title('age and bmi')
plt.grid()
plt.show()

plt.boxplot([data['children']],labels=['children'],vert = False)
plt.xticks(np.arange(0,6,0.5))
plt.title('children')
plt.grid()
plt.show()

plt.boxplot([data['charges']],labels=['charges'],vert = False)
plt.title('charges')
plt.grid()
plt.show()
```
